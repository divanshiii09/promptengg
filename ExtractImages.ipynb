```python
import requests
from bs4 import BeautifulSoup
from PIL import Image, UnidentifiedImageError
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from urllib.parse import urljoin
```

```python
page_url = input("Enter a webpage URL: ")
```

```python
try:
    response = requests.get(page_url)
    response.raise_for_status()
    soup = BeautifulSoup(response.content, 'html.parser')

    images = soup.find_all('img')

    if not images:
        print("No images found on the page.")
    else:
        for i, img_tag in enumerate(images):
            img_src = img_tag.get('src')
            if not img_src:
                continue

            img_url = urljoin(page_url, img_src)

            if img_url.lower().endswith('.svg'):
                print(f"‚è© Skipping SVG: {img_url}")
                continue

            print(f"\nüñºÔ∏è Attempting to load image {i+1}: {img_url}")

            try:
                img_response = requests.get(img_url)
                img_response.raise_for_status()

                img = Image.open(BytesIO(img_response.content))
                width, height = img.size
                fmt = img.format

                plt.figure(figsize=(4, 4))
                plt.imshow(img)
                plt.axis('off')
                plt.title(f"Image {i+1}")
                plt.show()

                print("üìÑ Metadata:")
                print(f"üîó URL     : {img_url}")
                print(f"üìê Size    : {width} x {height}")
                print(f"üì¶ Format  : {fmt}")
                print("-" * 50)

                shown += 1
                if shown >= 15:
                    break
            except (UnidentifiedImageError, Exception) as e:
                print(f"‚ùå Skipping image: {e}")
        if shown == 0:
            print("üòï No usable images could be loaded.")

except Exception as e:
    print(f"‚ùå Failed to fetch webpage: {e}")
```
