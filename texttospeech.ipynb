```python
from gtts import gTTS
from IPython.display import Audio, display, Markdown
import random
import string
from datetime import datetime
```

```python
text = input("Type something: ")
```

```python
def unique_filename():
    time_stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    rand_str = ''.join(random.choices(string.ascii_letters, k=5))
    return f"output_{time_stamp}_{rand_str}.mp3"

filename = unique_filename()
```

```python

tts = gTTS(text=text, lang='en', tld='ie') 
# tld='com'     ‚Üí American English (Default)
# tld='co.uk'   ‚Üí British English
# tld='com.au'  ‚Üí Australian English
# tld='co.in'   ‚Üí Indian English
# tld='ie'      ‚Üí Irish English
# tld='co.za'   ‚Üí South African English
tts.save(filename)
```

```python
display(Markdown(f"**üìù You typed:** `{text}`"))
display(Audio(filename))
```
